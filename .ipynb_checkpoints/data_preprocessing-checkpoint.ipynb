```python
import pandas as pd
```

```python
## Load the dataset

data = pd.read_csv('bank_marketing_data.csv', sep=';')
```

```python
# Display the few rows

data.head()
```

Data Preprocessing

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Remove duplicates if any
if duplicates > 0:
    data = data.drop_duplicates()
    print(f"Duplicates removed. New number of rows: {data.shape[0]}")

# Check for null values
null_values = data.isnull().sum()
print("\nNull values in each column:")
print(null_values[null_values > 0])

# Remove rows with null values
if null_values.sum() > 0:
    data.dropna(inplace=True)
    print(f"Rows with null values removed. New number of rows: {data.shape[0]}")

# Display data types and basic statistics
print("\nData Types and Basic Statistics:")
print(data.info())
print(data.describe())

# Save the cleaned data to a CSV file
data.to_csv('Bank_Marketing_data_cleaned.csv', index=False)
print("\nCleaned data saved as 'Bank_Marketing_data_cleaned.csv'.")
```
